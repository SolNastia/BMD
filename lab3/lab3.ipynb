```python
import seaborn as sns
%matplotlib inline
```

1. Тестові дані

```python
diamonds = sns.load_dataset('diamonds')
diamonds.head() 
```

2. Візуалізація даних

- *Графіки для аналізу кількісних даних*

```python
sns.distplot(diamonds['carat'])
```

```python
sns.jointplot(x='carat', y='table', data=diamonds, kind ='resid', color='y') 
```

```python
sns.pairplot(diamonds, hue='cut')  
```

- *Графіки для аналізу категоріальних даних*

```python
import numpy as np
sns.barplot(x='cut', y='price', data=diamonds, estimator=np.median)
```

```python
sns.boxplot(x='color', y='price', data=diamonds, palette="rainbow")
```

```python
sns.stripplot(x='color', y='depth', data=diamonds) 
```

- *Графіки для аналізу взаємозв'язку даних*

```python
sns.heatmap(diamonds.corr())
```

- *Графіки для відображення даних у вигляді сітки*

```python
import matplotlib.pyplot as plt
g = sns.FacetGrid(diamonds, col='cut', row='color')
g = g.map(plt.hist, 'price')
```
