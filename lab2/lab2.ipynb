1. Основи побудови графіків

```python
import numpy as np
x = np.linspace(0,5,11)
y = x ** 2
```

```python
x
```

```python
y
```

2. Побудова графіків  
2.1 Функціональний метод

```python
plt.plot(x,y)
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Graph') 
```

```python
plt.subplot(1,2,1)
plt.plot(x,y, 'g')
plt.subplot(1,2,2)
plt.plot(y,x, 'm')
```

2.2 Об'єктно-орієнтований метод

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x,y)
axes.set_xlabel('X')
axes.set_ylabel('Y')
axes.set_title('Graph') 
```

```python
fig = plt.figure()

axes1 = fig.add_axes([0, 0, 1, 1])
axes2 = fig.add_axes([0.1, 0.5, 0.5, 0.4])

axes1.plot(x,y, 'g')
axes1.set_xlabel('X')
axes1.set_ylabel('Y')
axes1.set_title('Graph')

axes2.plot(y,x, 'm')
axes2.set_xlabel('X')
axes2.set_ylabel('Y')
axes2.set_title('Graph')
```

```python
fig, axes = plt.subplots(nrows=1, ncols=2)

axes[0].plot(x, y, 'g')
axes[0].set_xlabel('X')
axes[0].set_ylabel('Y')

axes[1].plot(y, x, 'm')
axes[1].set_xlabel('X')
axes[1].set_ylabel('Y')
```

```python
fig, axes = plt.subplots(nrows=4, ncols=4)
plt.tight_layout() 
```

```python
fig = plt.figure()

axes = fig.add_axes([0,0,1,1])

axes.plot(x, x**3, label='x^3')
axes.plot(x, x**2, label='x^2')
axes.legend()
```

```python
fig = plt.figure(figsize=(8,4), dpi = 100)
```

```python
fig, axes = plt.subplots()
axes.plot(x, x**3, color ='y', marker='*', markersize = 8, markerfacecolor='b', lw = 1, ls = '-')
```

```python
fig.savefig("fig.pdf")
```
